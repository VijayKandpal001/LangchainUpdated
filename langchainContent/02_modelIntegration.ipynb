### Streaming and Batch

Most model can stream their output content while it is being generated. By this streaming improves the user experience. 
Calling stream() returns an iterator that yeilds output chunks as they are produced. You can use loop to process each chunk in real time. 

```python
modelgemini.invoke("Write me a 200 word essay on the history of Rajasthan").content
```

```python
# modelgemini.stream("Write me a 200 word essay on the history of Rajasthan")
for chunk in modelgemini.stream("Tell me about current affairs between the countries of the world"):
    print(chunk.text, end="", flush=True)
```

#### Batch

Batching is a collection of independent request to a model that can significantly improves the performance and reduce costs, as the processing can be done in parallel.

```python
import re

responses=modelgroq.batch([
    "What is the capital of Rajasthan?",
    "What is the capital of India?",
    "What is the capital of USA?"
], 
config={
    'max_concurrency': 3 # maximum 3 parallel requests
    }
)
for response in responses:
    clean_content = re.sub(r'<think>.*?</think>', '', response.content, flags=re.DOTALL).strip()
    print(clean_content)
```
