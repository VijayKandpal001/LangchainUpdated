## Tools

Model can request the tools to perform tasks fetching data from the database, searching web, running code.Tools are pairing of:

- A schema, inclduing the name of the tool, description, and/or argument definitions(often JSON schema)
- A function or coroutine to execute.

```python
import os
from langchain.chat_models import init_chat_model

os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model=init_chat_model("groq:qwen/qwen3-32b")    
response=model.invoke("Hi there")
response
```

```python
from langchain.agents import create_agent
from langchain.tools import tool
@tool
def getAQI(city: str)-> str:
    """Get the Air Quality Index (AQI) for a given city."""
    return f"The AQI for {city} is 150, which is considered unhealthy."

model_with_tools=model.bind_tools([getAQI])
```

```python
response= model_with_tools.invoke("What is the AQI for Delhi?")
# print(response)
for tool_call in response.tool_calls:
    print(f"Tool Name: {tool_call['name']}")
    print(f"Arguments: {tool_call['args']}")
```
