### Messages
Messages are the fundamental unit of context for models in LangChain. They represent the input and output of models, carrying both the content and metadata needed to represent the state of a conversation when interacting with an LLM. Messages are objects that contain:

- Role - Identifies the message type (e.g. system, user)
- Content - Represents the actual content of the message (like text, images, audio, documents, etc.)
- Metadata - Optional fields such as response information, message IDs, and token usage

LangChain provides a standard message type that works across all model providers, ensuring consistent behavior regardless of the model being called.

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

model=init_chat_model("groq:qwen/qwen3-32b")
model.invoke("Where is BLM Academy Higher Secondary School located in Uttarakhand?").content
```

### Message Prompts

Alternatively, you can pass in a list of messages to the model by providing the list of model objects

Message types:
+ System messages: tell the model how to behave and provide the context for the interactions
+ Human messages: Represent the user input and the interactions with the model 
+ AI messages: Responses generated by the model including the text context, toolscalls and metadata
+ Tool messages: Represents the outputs of the tool call


### System Message:
A SystemMessage represents the initial set of instructions that primes the model's behaviour. You can use the system message to set the tone, define the role of the model, and establish the guidelines for the responses


### Human Message:
A HumanMessage represents user input and interactions. They can contain text ,images, audio, files and any other kind of multimodel content.

### AI Message:
An AIMessage represents the output of the model invocation. They can include the multimodal data, tool calls, provides specific metadata.

### Tool Messages:
For model that supports tool calling, AIMessage can contains the toolcalls. ToolMessages are used to pass the result of the single tool execution back to the model.

```python
from langchain.messages import SystemMessage, HumanMessage, AIMessage, ToolMessage
message=[
    SystemMessage(content="You are an indian poet. You write poems in hinglish language but takes words from hindia and urdu language and write poems in hinglish language. You are a very good poet and you have won many awards for your poetry."),
    HumanMessage(content="Write a poem on love in hinglish language containing urdu words also."),
]

# response=model.invoke(message)
# print(response.content)
response=geminimodel.invoke(message)
print(response.content)
```

```python
ai_msg=AIMessage(
    content="I would surely help you with your problem.",
)
message=[
    SystemMessage("You are a helpful assistant."),
    HumanMessage(content="Can you help me?"),
    ai_msg,
    HumanMessage(content="what is 15/3?"),
]
mathresponse=model.invoke(message)
print(mathresponse.content)
```

```python
# Tool Message Example
ai_msg=AIMessage(
    content=[],
    tool_calls=[{
        "name": "getAQI",
        "args": {"city": "Delhi"},
        "id":"ABC123"
    }]
)

aqi_result="600 AQI for Delhi"
tool_msg=ToolMessage(
    content=aqi_result,
    tool_call_id="ABC123"
)

message=[
    SystemMessage("You are a helpful assistant."),
    HumanMessage(content="What is the AQI for Delhi?"),
    ai_msg,
    tool_msg
]
aqians=model.invoke(message)
print(aqians.content)
```
